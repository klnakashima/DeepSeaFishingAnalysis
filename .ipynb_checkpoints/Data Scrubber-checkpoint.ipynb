```python
# IMPORTS
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("fishCounts.csv",index_col=0)
df.head()
```

```python
countsdf = df[["Fish Count"]].copy()
countsdf.columns = ["fish"]
countsdf.head()
```

```python
s = countsdf.fish.str.split(", ", expand=True).stack()
s.head()
```

```python
s = s.str.extract('(\d+)(\D+)')
s.head()
```

```python
s.columns = ['num','species']
s.species = s.species.str.strip()
s = s.reindex(['species','num'],axis=1)
s.head()
```

```python
s.index.names = ['a','b']
s.head()
```

```python
s_flat = s.reset_index()
s_flat.head()
```

```python
s_reindexed = s_flat.set_index(['a','species'])
s_reindexed.head()
```

```python
s_reindexed = s_reindexed.drop(columns='b')
s_reindexed.head()
```

```python
#pd.set_option('display.max_columns',None)
s_reindexed = s_reindexed.unstack(1)
s_reindexed.head()
```

```python
s_reindexed.columns = s_reindexed.columns.droplevel(0)
s_reindexed.head()
```

```python
s_reset = s_reindexed.drop(columns=np.nan)
s_reset.head()
```

```python
s_reset = s_reset.fillna(0)
s_reset.head()
```

```python
fulldf = df.join(s_reset, how='outer')
fulldf = fulldf.drop(['Fish Count'], axis=1)
fulldf.head()
```

```python
fulldf.iloc[:,4:] = fulldf.iloc[:,4:].astype(int)
fulldf['Date'] = pd.to_datetime(fulldf['Date'])
fulldf.info()
```

```python
daveysdf = fulldf[fulldf.Landing=="daveys-locker"]
danadf =  fulldf[fulldf.Landing=="dana-wharf-sportfishing"]
daveysdf.to_csv("cleanData-daveys.csv")
danadf.to_csv("cleanData-dana.csv")
```
